```python
data_01 = np.zeros((3,3))
data_01
```

```python
# Chia data thành 3 nhóm tuổi: nhóm trẻ, trung niên và người cao tuổi
Low_01 = top_1[top_1['Age'] < 30]
Mid_01 = top_1[(top_1['Age'] >= 30) & (data['Age'] < 60)]
High_01 = top_1[top_1['Age'] >= 60]
```

```python
age = []
age.append(Low_01)
age.append(Mid_01)
age.append(High_01)
```

```python
for j in range(3):
    data_01[0][j] = age[j][age[j]['Rating'] > 3]['Clothing ID'].count()
    data_01[1][j] = age[j][age[j]['Rating'] == 3]['Clothing ID'].count()
    data_01[2][j] = age[j][age[j]['Rating'] < 3]['Clothing ID'].count()
```

```python
X = data_01.sum(axis=0)
X
```

```python
for j in range(3):
    data_01[:,j] = np.round(data_01[:,j] / X[j],2)
```

```python
# Plot the Marimekko chart

fig, ax = plt.subplots(1)
x_label = ['Low','Mid','High']

width = [round(i/sum(X),2) for i in X]
adjusted_x, temp = [0], 0
for i in width[:-1]:
    temp += i
    adjusted_x.append(temp)

plt.bar(adjusted_x, data_01[0], width=width, align='edge', edgecolor='white', color='#06d6a0', label='Good Rating')
plt.bar(adjusted_x, data_01[1], bottom=data_01[0], width=width, align='edge', edgecolor='white', color='#ffd166', label='Averange Rating')
plt.bar(adjusted_x, data_01[2], bottom=data_01[1] + data_01[0], width=width, align='edge', edgecolor='white', color='#ef476f', label='Bad Rating')

plt.ylim(0,1)
plt.xlim(0,1)

ax.set_yticks([0, 0.25, 0.5, 0.75, 1])
ax.set_yticklabels(['0%', '25%', '50%', '75%', '100%'])
ax.set_xticks([0, 0.25, 0.5, 0.75, 1])
ax.set_xticklabels(['0%', '25%', '50%', '75%', '100%'])

axy = ax.twiny()
axy.set_xticks([(width[i]/2)+ v for i, v in enumerate(adjusted_x)])
axy.set_xticklabels(x_label, fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
axy.spines['top'].set_visible(False)
axy.spines['right'].set_visible(False)

ax.legend()
plt.show()
```

- Lượt rating tích cực của sản phẩm này rất cao và tập trung chủ yếu vào độ cuổi Mid -> Từ histogram phía trên cũng đã giải thích được nguyên nhân nay là do độ tuổi tiêu dùng sản phẩm này nhiều nhất là từ 30 -50 tuổi.

- Giới trẻ và người cao tuổi thường có đánh giá tích cực nhiều hơn.
